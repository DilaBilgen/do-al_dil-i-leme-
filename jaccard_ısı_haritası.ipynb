```python
model_results = {
    "tfidf_lemmatized": {"doc1", "doc2", "doc3", "doc4", "doc5"},
    "tfidf_stemmed": {"doc3", "doc4", "doc5", "doc6", "doc7"},
    "w2v_cbow_100_win5": {"doc2", "doc3", "doc6", "doc8", "doc9"},
    "w2v_sg_100_win5": {"doc3", "doc4", "doc9", "doc10", "doc11"},}


def jaccard_similarity(set1, set2):
    return round(len(set1 & set2) / len(set1 | set2), 2)
    
import pandas as pd

model_names = list(model_results.keys())
jaccard_matrix = pd.DataFrame(index=model_names, columns=model_names)

for m1 in model_names:
    for m2 in model_names:
        jaccard_matrix.loc[m1, m2] = jaccard_similarity(model_results[m1], model_results[m2])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

jaccard_matrix = jaccard_matrix.astype(float)

plt.figure(figsize=(12, 8))
sns.heatmap(jaccard_matrix, annot=True, cmap="Blues", linewidths=0.5)
plt.title("Jaccard Benzerlik Is Haritas")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
